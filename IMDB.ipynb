### 2. What are the top 3 movies with highest nett profit in 2013? 

```python
movie13 = imdb[imdb.title_year == 2013]
net_sort = movie13.sort_values("net_USD", ascending=False)
net_sort.head(3)
```

### 3. What are the top 3 movies with highest gross profit in 2009? 

```python
movie09 = imdb[imdb.title_year == 2009]
gross09 = movie09.sort_values('gross_USD', ascending=False)
gross09.head(3)
```

### 4. How many non-english movies are in the dataset? 

```python
non_english = imdb[imdb.language != 'English']
#total = non_english['movie_ID'].count()
print ('Non-English movies in the dataset are',len(non_english))
```

### 5. What are top 3 languages in the dataset aside English (based on number of movies)?

```python
#movie_country = imdb[imdb.country != 'English']
#net_sort = movie_country.sort_values('net_USD')
#net_sort.head(3)

non_english1 = imdb[imdb.language != 'English']
grouping = non_english1.groupby('language')
total = grouping['movie_ID'].count()
#print ('Non-English movies in the dataset are',total)
non_sort = total.sort_values(ascending = False)
non_sort.head(3)
```

### 6. What are the top 3 countries with the highest nett profit in all dataset times? 

```python
#movie_all = imdb.loc[(imdb.title_year >= 1927) & (imdb.title_year <= 2016)]
#net_all = movie_all.sort_values('net_USD')
#net_all.head(3)

movie_all = imdb.groupby('country')
movie_all_count = movie_all['net_USD'].sum()
sort_all = movie_all_count.sort_values(ascending=False)
sort_all.head(3)
```

### 7. What are the average budget of all movies in dataset in 2013? 

```python
avg13 = imdb[(imdb.title_year == 2013)]
avg131 = avg13.budget_USD.sum()

average  = avg131 / len(avg13)
average
```

### 8. What are the average budget of all movies in dataset in 2014? 

```python
avg14 = imdb[(imdb.title_year == 2014)]
avg141 = avg14.budget_USD.sum()

average1  = avg141 / len(avg14)
average1
```

### 9. Which one is higher, average budget of all movies in dataset in 2014 or 2013? 

```python
hasil = max(average, average1, key=lambda x:int(x))
print('rata-rata tertinggi diperoleh pada angka', hasil)
```
